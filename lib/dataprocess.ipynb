```python
# Define a function to map the old categories to new, broader ones
def categorize_income(income_str):
    if 'Less than $10,000' in income_str:
        return 'Less than $10,000'
    elif any(x in income_str for x in ['$10,000 to $19,999', '$20,000 to $29,999', '$30,000 to $39,999', '$40,000 to $49,999','$10,000 to $14,999','$15,000 to $24,999','$25,000 to $34,999','$35,000 to $49,999']):
        return '$10,000 to $49,999'
    elif any(x in income_str for x in ['$50,000 to $59,999','60,000 to $99,999','$50,000 to $74,999','$75,000 to $99,999']):
        return '$50,000 to $99,999'
    elif '$100,000 to $149,999' in income_str:
        return '$100,000 to $149,999'
    elif any(x in income_str for x in ['$150,000 or more','$150,000 to $199,999','$200,000 or more']):
        return '$150,000 or more'
    else:
        return income_str  # Return the original string if it doesn't match any category

# Apply the categorization function to the income_category column
df21['broad_income'] = df21['income'].apply(categorize_income)
df22['broad_income'] = df22['income'].apply(categorize_income)
df23['broad_income'] = df23['income'].apply(categorize_income)
```

```python
# Define a function to map the old categories to new, broader ones
def categorize_age(age_str):
    if any(x in age_str for x in ['18-59','18-19','20-29','18-29','30-39','40-49','50-59']):
        return '18-59'
    elif any(x in age_str for x in ['60-69','70-79','Over 80','60+']):
        return '60+'
    else:
        return age_str  # Return the original string if it doesn't match any category

# Apply the categorization function to the income_category column
df21['broad_age'] = df21['age'].apply(categorize_age)
df22['broad_age'] = df22['age'].apply(categorize_age)
df23['broad_age'] = df23['age'].apply(categorize_age)
```

```python
def categorize_race(race_str):
    if any(x in race_str for x in ['Native Hawaiian or Other Pacific Islander']):
        return 'Native Hawaiian or Pacific Islander'
    else:
        return race_str  # Return the original string if it doesn't match any category

df21['broad_race'] = df21['race'].apply(categorize_race)
df22['broad_race'] = df22['race'].apply(categorize_race)
df23['broad_race'] = df23['race'].apply(categorize_race)
```

```python
df21['broad_edu'] = df21['education']
df22['broad_edu'] = df22['education']
df23['broad_edu'] = df23['education']
```

```python
# Define a function to map the old categories to new, broader ones
def categorize_awareness(awareness_str):
    if any(x in awareness_str for x in ['I have been prepared for the last year']):
        return 'I have been prepared for LESS than a year'
    elif any(x in awareness_str for x in ['Prefer not to answer']):
        return "Don't know"
    else:
        return awareness_str  # Return the original string if it doesn't match any category

# Apply the categorization function to the income_category column
df21['broad_awareness'] = df21['awareness level'].apply(categorize_awareness)
df22['broad_awareness'] = df22['awareness level'].apply(categorize_awareness)
df23['broad_awareness'] = df23['awareness level'].apply(categorize_awareness)
```
